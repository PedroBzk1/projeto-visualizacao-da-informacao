Projeto Vizualização da Informação

Instalando as Bibliotecas que vamos utilizar

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Importanto meu dataset com pandas

```python
dados = pd.read_csv ('countries-by-intentional-homicide-rate.csv')
```

```python
dados['Location']
dados['Count']
```

```python
dados['Location'].unique().size
```

```python
dados['Region'].unique()
```

Dando tamanho ao gráfico e deixando ele correto

```python
plt.figure(figsize=(7,5)) ##para dar tamanho no grafico
plt.bar(dados['Region'], dados['Count']);
plt.xticks(rotation=30); ##para rodar o nome das legendas
plt.ylabel( 'taxa de homicídios intencionais'); ##dar nome no eixo Y
plt.title('Taxa de Homicídio intencional por Continentes (2006-2021)');
plt.grid();
```

GRÁFICO 2

```python
import plotly.express as px
fig = px.treemap(dados, path=['Region', 'Subregion'], values='Count');
fig.show()
```
